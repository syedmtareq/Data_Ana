# Data cleaning 

```python
Air_Quality.isnull().sum()
```

```python
Air_Quality[Air_Quality['PM 2.5'].isnull()]
```

# Data update

```python
Air_Quality_Updated = Air_Quality.interpolate()
Air_Quality_Updated 
```

# Data normalization 

```python
Air_Quality_Updated = Air_Quality_Updated.apply(lambda x: (x-x.min(axis=0)) / (x.max(axis=0)-x.min(axis=0)))
```
